```python
red_wine['Color'] = 'Red'
```

```python
white_wine.shape
```

```python
merge_data = pd.concat([white_wine,red_wine], axis = 0)
```

```python
merge_data.tail(3000)
```

```python
bins = [0, 6, 10]
names = ['Bad', 'Good']

merge_data['quality'] = pd.cut(merge_data['quality'], bins, labels=names)
```

```python
merge_data
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
cols_numeric = merge_data.select_dtypes(include=numerics).columns.tolist()
```

```python
cols_numeric
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
merge_data[cols_numeric]=scaler.fit_transform(merge_data[cols_numeric])
```

```python
merge_data
```

```python
merge_data['Color'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

quality_encoder = LabelEncoder()
merge_data["quality"] = quality_encoder.fit_transform(merge_data["quality"])
```

```python
from sklearn.preprocessing import LabelEncoder

color_encoder = LabelEncoder()
merge_data["Color"] = color_encoder.fit_transform(merge_data["Color"])
```

```python
import matplotlib
import matplotlib.pyplot as plt
plt.matshow(merge_data.corr(method='spearman'))
plt.xticks(range(len(merge_data.columns)), merge_data.columns)
plt.yticks(range(len(merge_data.columns)), merge_data.columns)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
corr=merge_data.corr(method='spearman')
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
